```python
import math
import numpy as np
import matplotlib.pyplot as plt
import sys
```

```python
%matplotlib notebook
```

```python
def f(x, y):
    return x**4.0 - 4.0*x*y + y**4.0
```

```python
x_min, x_max = 0.5, 2.5
y_min, y_max = -5.0, 4.0
X, Y = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
Z = np.empty(X.shape)

for u in range(Z.shape[0]):
    for v in range(Z.shape[1]):
        Z[u, v] = f(X[u, v], Y[u, v])
```

```python
plt.figure()
cs = plt.contour(X, Y, Z, 20)
plt.clabel(cs, inline=1, fontsize=10)
plt.title('f(x, y) contour')
plt.xlabel('x'); plt.ylabel('y')
# plt.gcf().savefig('contour_21.pdf')
```

```python
X0 = 0
X = -1.0
X2 = 1.0

y_min = 2
y_max = -2
y_star = [-1, 1]
Y = np.linspace(y_min, y_max, 100)
Z0 = np.empty(Y.shape)
Z = np.empty(Y.shape)
Z2 = np.empty(Y.shape)
for u in range(Y.size):
    Z0[u] = f(X0, Y[u])

for u in range(Y.size):
    Z[u] = f(X, Y[u])

for u in range(Y.size):
    Z2[u] = f(X2, Y[u])
    
plt.figure()
plt.plot(Y, Z0)
plt.plot(Y, Z)
plt.plot(Y, Z2)
z_min = plt.gca().get_ylim()[0]

plt.plot([0, 0], [z_min, f(X0, 0)], color='gray', linestyle='dashed')
plt.plot([1, 1], [z_min, f(X2, 1)], color='gray', linestyle='dashed')
plt.plot([-1, -1], [z_min, f(X, -1)], color='gray', linestyle='dashed')
plt.legend(["x = 0", "x = 1", "x = 2"], loc ="lower right") 
plt.title("f(x, y)")
plt.xlabel('y')
# plt.gcf().savefig('solution_212.pdf')
```
