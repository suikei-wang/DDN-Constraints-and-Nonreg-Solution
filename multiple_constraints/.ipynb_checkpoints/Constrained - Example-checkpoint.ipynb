```python
import math
import numpy as np
import matplotlib.pyplot as plt
import sys
from matplotlib.patches import Circle, Ellipse, Rectangle, Polygon
from matplotlib.collections import PatchCollection
```

```python
%matplotlib notebook
```

```python
x1 = np.linspace(-5, 1, 100)
x2 = np.linspace(-5, 10, 100)
x3 = np.linspace(0, 10, 100)
y0 = 7
y1 = -4
y2 = 0
y3 = 7
y4 = np.sqrt(x3)
y5 = -5
```

```python
fig, ax = plt.subplots()
plt.vlines(1, -3, 5, color = 'green')
plt.hlines(0, -3, 5, color = 'blue')
plt.plot(x3, y4, color='black')
plt.legend(["$y^2 - x = 0$", "$x - 1 = 0$", "$y = 0$"], loc ="lower right") 
axes = plt.gca()
axes.set_xlim([-3,3])
axes.set_ylim([-3,3])


ax.fill_between(x1, y0, y1, alpha=0.2, color='red')
ax.fill_between(x2, y2, y3, alpha=0.2, color='red')
ax.fill_between(x3, y4, y5, alpha = 0.2, color='red')

ax.set_xlabel('x')
ax.set_ylabel('y')
# plt.gcf().savefig('solution_22.pdf')
```
